### Dense Layer Class

So far I have only used what is called a dense or fully-connected layer. These layers are commonly referred to as "dense" layers.

```python
class Layer_Dense:
    # layer initialization constructor
    def __init__(self, n_inputs, n_neurons):
        # initialize weights and biases
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        # Inputs, neurons instead of neurons, inputs because we do not want to transpose the weights on every forward pass
        self.biases = np.zeros((1, n_neurons))
        # 1D array of zeros representing the biases for each neuron

    # forward pass
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
        
```

Making use of this new class instead of hardcoded calculations. First let us generate a dataset using the nnfs library by sentdex.

```python
from nnfs.datasets import spiral_data
import matplotlib.pyplot as plt

X, y = spiral_data(samples=100, classes=3)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='brg')
plt.show()
```

```python
# Creating a dense layer with 2 input features and 3 neurons
dense1 = Layer_Dense(2, 3)

# Performing a forward pass of our training data thorugh this layer
dense1.forward(X)
#Output of first dense layer (few samples)
print(dense1.output[:5])
```
